```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# CSVファイルの読み込み
file_path_1 = 'ants_1.csv'
file_path_2 = 'ants_2.csv'
file_path_3 = 'ants_3.csv'
file_path_4 = 'ants_4.csv'
file_path_5 = 'ants_5.csv'
file_path_6 = 'ants_6.csv'
file_path_7 = 'ants_7.csv'
file_path_8 = 'ants_8.csv'
file_path_9 = 'ants_9.csv'
file_path_10 = 'ants_10.csv'
file_path_11 = 'ants_11.csv'
file_path_12 = 'ants_12.csv'
file_path_12_2 = 'ants_12_2.csv'
file_path_13 = 'ants_13.csv'
file_path_14 = 'ants_14.csv'

data_1 = pd.read_csv(file_path_1)
data_2 = pd.read_csv(file_path_2)
data_3 = pd.read_csv(file_path_3)
data_4 = pd.read_csv(file_path_4)
data_5 = pd.read_csv(file_path_5)
data_6 = pd.read_csv(file_path_6)
data_7 = pd.read_csv(file_path_7)
data_8 = pd.read_csv(file_path_8)
data_9 = pd.read_csv(file_path_9)
data_10 = pd.read_csv(file_path_10)
data_11 = pd.read_csv(file_path_11)
data_12 = pd.read_csv(file_path_12)
data_12_2 = pd.read_csv(file_path_12_2)
data_13 = pd.read_csv(file_path_13)
data_14 = pd.read_csv(file_path_14)
```

```python
# グラフの描画
plt.plot(data_1['time'], data_1['wait'], label='ants_1')
plt.plot(data_2['time'], data_2['wait'], label='ants_2')
plt.plot(data_3['time'], data_3['wait'], label='ants_3')
plt.title('Wait vs Time')
plt.xlabel('Time')
plt.ylabel('Wait')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# グラフの描画
plt.plot(data_1['time'], data_1['food'], label='ants_1')
plt.plot(data_2['time'], data_2['food'], label='ants_2')
plt.plot(data_3['time'], data_3['food'], label='ants_3')
plt.title('Food vs Time')
plt.xlabel('Time')
plt.ylabel('Food')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# グラフの描画
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # 1行2列のサブプロットを作成

# グラフ1: Wait vs Time
ax1.plot(data_1['time'], data_1['wait'], label='ants_1')
ax1.plot(data_2['time'], data_2['wait'], label='ants_2')
ax1.plot(data_3['time'], data_3['wait'], label='ants_3')
ax1.set_title('Wait vs Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Wait')
ax1.legend()
ax1.grid(True)

# グラフ2: Food vs Time
ax2.plot(data_1['time'], data_1['food'], label='ants_1')
ax2.plot(data_2['time'], data_2['food'], label='ants_2')
ax2.plot(data_3['time'], data_3['food'], label='ants_3')
ax2.set_title('Food vs Time')
ax2.set_xlabel('Time')
ax2.set_ylabel('Food')
ax2.legend()
ax2.grid(True)

# グラフ間のスペースを調整
plt.tight_layout()

plt.show()
```

```python
# グラフの描画
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))  # 1行2列のサブプロットを作成

# グラフ1: Wait vs Time
ax1.plot(data_7['time'], data_7['wait'], label='ants_7')
ax1.plot(data_8['time'], data_8['wait'], label='ants_8')
ax1.plot(data_9['time'], data_9['wait'], label='ants_9')

# 縦軸の目盛りを0から10の感覚で設定
y_ticks = np.arange(0, max(max(data_7['wait']), max(data_8['wait']), max(data_9['wait'])) + 15, 10)
ax1.set_yticks(y_ticks)
# 0の少し下に余裕を持たせる
ax1.set_ylim(-2, max(max(data_7['wait']), max(data_8['wait']), max(data_9['wait'])) + 10)

ax1.set_title('Wait vs Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('Wait')
ax1.legend()
ax1.grid(True)

# グラフ2: Food vs Time
ax2.plot(data_7['time'], data_7['food'], label='ants_7')
ax2.plot(data_8['time'], data_8['food'], label='ants_8')
ax2.plot(data_9['time'], data_9['food'], label='ants_9')
ax2.set_title('Food vs Time')
ax2.set_xlabel('Time')
ax2.set_ylabel('Food')
ax2.legend()
ax2.grid(True)

# グラフ間のスペースを調整
plt.tight_layout()

plt.show()
```
